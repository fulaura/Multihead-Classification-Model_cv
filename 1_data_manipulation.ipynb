```python
df['emotion'] = df['emotion'].apply(lambda x: 2 if x==3 else x)
df['emotion'] = df['emotion'].apply(lambda x: x-1 if x>2 else x)
df['emotion']-=1
df['emotion'].value_counts()
df['emotion'] = df['emotion'].apply(lambda x: 3 if x==4 else x)
df['emotion'] = df['emotion'].apply(lambda x: x-1 if x==5 else x)
df['emotion'].value_counts()
```
